```python
PATH = ['./test_data/edgelist_days/edgelist_w1_2022-02-01.txt',
       './test_data/edgelist_days/edgelist_w1_2022-02-02.txt',
       './test_data/edgelist_days/edgelist_w1_2022-02-03.txt']
```

```python
all_nodes = set()
for path in PATH:
    df = pd.read_csv(path, sep ='\t', header = None)
    df.columns = ['node1','node2','weight']
    unique_nodes = set(df['node1'].unique())|set(df['node2'].unique())
    all_nodes = all_nodes|unique_nodes
```

```python
mylist = ["A", "B"]
print(random.choice(mylist))
```

```python
dict_nodes_community = {}
for node in all_nodes:
    dict_nodes_community[str(node)]=random.choice(mylist)
```

```python
# Open a file and use dump() 
with open('com_of_user_days.pkl', 'wb') as file: 
      
    # A new file will be created 
    pickle.dump(dict_nodes_community, file)
```
